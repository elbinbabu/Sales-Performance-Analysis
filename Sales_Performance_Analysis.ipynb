```python
import pandas as pd
df = pd.read_csv("train.csv")
df.head()
```

```python
df.shape
df.info()
df.describe()
```

```python
sales_region = df.groupby("Region")["Sales"].sum()
sales_region
```

```python
sales_category = df.groupby("Category")["Sales"].sum()
sales_category
```

```python
import matplotlib.pyplot as plt
sales_region.plot(kind="bar")
plt.title("Total Sales by Region")
plt.ylabel("Sales")
plt.show()
```
